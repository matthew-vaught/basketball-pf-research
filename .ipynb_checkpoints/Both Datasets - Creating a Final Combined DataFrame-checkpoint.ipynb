```python
# Load in the 2014 basketball-reference.com data
bball_ref_data_2014 = pd.read_csv('bball-ref_data_2014.csv')
bball_ref_data_2014
```

```python
# Load in the 2024 basketball-reference.com data
bball_ref_data_2024 = pd.read_csv('bball-ref_data_2024.csv')
bball_ref_data_2024
```

```python
# load in the 2014 nba.com data
nba_data_2014 = pd.read_csv('nba_data_2014.csv')
nba_data_2014
```

```python
# load in the 2024 nba.com data
nba_data_2024 = pd.read_csv('nba_data_2024.csv')
nba_data_2024
```

```python
import difflib

# Drop unneccessary columns in the nba.com dataframes
col_to_drop = ['Unnamed: 0', 'PTS%', 'FT%', 'Contested OREB', 'Contested OREB%', 'OREB Chances', 'Deferred OREB Chances', 'Adjusted OREB Chance%', 'AVG OREB Distance', 'Contested DREB', 'Contested DREB%', 'DREB Chances', 'Deferred DREB Chances', 'Adjusted DREB Chance%', 'AVG DREB Distance', 'DFGM', 'DFGA', 'DFG%']

for col in col_to_drop:
    close_matches = difflib.get_close_matches(col, nba_data_2014.columns, n=1)
    nba_data_2014.drop(columns=close_matches, inplace=True)
    nba_data_2024.drop(columns=close_matches, inplace=True)
```

```python
# Rename columns in the nba.com dataframes for better understanding
nba_data_2014.rename(columns = {'FGM': 'FGM - DRIVES', 'FGA': 'FGA - DRIVES', 'FG%': 'FG% - DRIVES', 'FTM': 'FTM - DRIVES', 'FTA': 'FTA - DRIVES', 'PASS': 'PASS - DRIVES', 'PASS%': 'PASS% - DRIVES', 'AST': 'AST - DRIVES', 'AST%': 'AST% - DRIVES', 'TO': 'TO - DRIVES', 'TOV%': 'TOV% - DRIVES', 'PF': 'PF - DRIVES', 'PF%': 'PF% - DRIVES'}, inplace=True)
nba_data_2024.rename(columns = {'FGM': 'FGM - DRIVES', 'FGA': 'FGA - DRIVES', 'FG%': 'FG% - DRIVES', 'FTM': 'FTM - DRIVES', 'FTA': 'FTA - DRIVES', 'PASS': 'PASS - DRIVES', 'PASS%': 'PASS% - DRIVES', 'AST': 'AST - DRIVES', 'AST%': 'AST% - DRIVES', 'TO': 'TO - DRIVES', 'TOV%': 'TOV% - DRIVES', 'PF': 'PF - DRIVES', 'PF%': 'PF% - DRIVES'}, inplace=True)
```

```python
# Adjusting some of the stats and removing the remaining unneccesary stat columns
nba_data_2014['FGA per drive'] = (nba_data_2014['FGA - DRIVES'] / nba_data_2014['DRIVES']).round(3)
nba_data_2014['FTA per drive'] = (nba_data_2014['FTA - DRIVES'] / nba_data_2014['DRIVES']).round(3)
nba_data_2014['AST per drive'] = (nba_data_2014['AST - DRIVES'] / nba_data_2014['DRIVES']).round(3)
nba_data_2014['TOV per drive'] = (nba_data_2014['TO - DRIVES'] / nba_data_2014['DRIVES']).round(3)

nba_data_2024['FGA per drive'] = (nba_data_2024['FGA - DRIVES'] / nba_data_2024['DRIVES']).round(3)
nba_data_2024['FTA per drive'] = (nba_data_2024['FTA - DRIVES'] / nba_data_2024['DRIVES']).round(3)
nba_data_2024['AST per drive'] = (nba_data_2024['AST - DRIVES'] / nba_data_2024['DRIVES']).round(3)
nba_data_2024['TOV per drive'] = (nba_data_2024['TO - DRIVES'] / nba_data_2024['DRIVES']).round(3)

more_col_to_drop = ['W', 'L', 'PTS PER ELBOW TOUCH', 'PTS PER POST TOUCH', 'PTS PER PAINT TOUCH', 'FGM - DRIVES', 'FGA - DRIVES', 'FG% - DRIVES', 'FTM - DRIVES', 'FTA - DRIVES', 'PASS - DRIVES', 'AST - DRIVES', 'TO - DRIVES', 'PF - DRIVES', 'PF% - DRIVES', 'eFG%']

for col in more_col_to_drop:
    close_matches = difflib.get_close_matches(col, nba_data_2014.columns, n=1)
    nba_data_2014.drop(columns=close_matches, inplace=True)
    nba_data_2024.drop(columns=close_matches, inplace=True)
```
