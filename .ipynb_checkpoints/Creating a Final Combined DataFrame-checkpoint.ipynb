```python
import pandas as pd
```

```python
# Load in the 2014 basketball-reference.com data
bball_ref_data_2014 = pd.read_csv('bball-ref_data_2014.csv')
bball_ref_data_2014
```

```python
# Load in the 2024 basketball-reference.com data
bball_ref_data_2024 = pd.read_csv('bball-ref_data_2024.csv')
bball_ref_data_2024
```

```python
# load in the 2014 nba.com data
nba_data_2014 = pd.read_csv('nba_data_2014.csv')
nba_data_2014
```

```python
# load in the 2024 nba.com data
nba_data_2024 = pd.read_csv('nba_data_2024.csv')
nba_data_2024
```

```python
# Drop unneccessary columns in the nba.com dataframes
nba_data_2014.drop(columns=['Unnamed: 0', 'PTS%', 'FT%'], inplace=True)
nba_data_2024.drop(columns=['Unnamed: 0', 'PTS%', 'FT%'], inplace=True)
```

```python
# Rename columns in the nba.com dataframes for better understanding
nba_data_2014.rename(columns = {'FGM': 'FGM-DRIVES', 'FGA': 'FGA-DRIVES', 'FG%': 'FG%-DRIVES', 'FTM': 'FTM-DRIVES', 'FTA': 'FTA-DRIVES', 'PASS': 'PASS-DRIVES', 'PASS%': 'PASS%-DRIVES', 'AST': 'AST-DRIVES', 'AST%': 'AST%-DRIVES', 'TO': 'TO-DRIVES', 'TOV%': 'TOV%-DRIVES', 'PF': 'PF-DRIVES', 'PF%': 'PF%-DRIVES'}, inplace=True)
nba_data_2024.rename(columns = {'FGM': 'FGM-DRIVES', 'FGA': 'FGA-DRIVES', 'FG%': 'FG%-DRIVES', 'FTM': 'FTM-DRIVES', 'FTA': 'FTA-DRIVES', 'PASS': 'PASS-DRIVES', 'PASS%': 'PASS%-DRIVES', 'AST': 'AST-DRIVES', 'AST%': 'AST%-DRIVES', 'TO': 'TO-DRIVES', 'TOV%': 'TOV%-DRIVES', 'PF': 'PF-DRIVES', 'PF%': 'PF%-DRIVES'}, inplace=True)
```
